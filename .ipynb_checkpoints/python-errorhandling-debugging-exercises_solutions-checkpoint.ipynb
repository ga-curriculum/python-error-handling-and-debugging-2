# Introduction to Error Handling and Debugging

Errors are an inevitable part of programming. Being able to understand, anticipate, and handle errors effectively is an essential skill for any Python developer. This notebook will help you learn how to:

- Identify common types of errors in Python
- Read and interpret error messages
- Handle errors gracefully using try/except blocks
- Debug your code effectively

### Types of Python Errors

Python errors generally fall into two categories:

1. **Syntax Errors**: These occur when Python can't understand your code due to incorrect syntax (like forgetting a colon, mismatched parentheses, etc.). Python detects these before your code runs.

2. **Exceptions**: These occur during execution when something unexpected happens (like trying to divide by zero, accessing a non-existent file, etc.). Python provides many built-in exception types to help identify what went wrong.

### Common Python Exceptions

Here are some of the most common exceptions you'll encounter:

- `SyntaxError`: Invalid Python syntax
- `TypeError`: Operation applied to inappropriate type
- `NameError`: Variable not defined
- `IndexError`: Index out of range
- `KeyError`: Key not found in dictionary
- `ValueError`: Correct type but inappropriate value
- `FileNotFoundError`: Trying to access a non-existent file
- `ZeroDivisionError`: Dividing by zero
- `ImportError`: Module could not be imported
- `IndentationError`: Incorrect indentation

### Reading Error Messages

Python error messages typically include:
- The type of error (e.g., `TypeError`, `ValueError`)
- A description of what went wrong
- A traceback showing where in your code the error occurred

Error messages are your friends! They provide valuable clues about what's wrong and where to look.

### Using Try/Except Blocks

Python's `try`/`except` blocks allow you to handle potential errors gracefully:

```python
try:
    # Code that might cause an exception
    result = 10 / 0
except ZeroDivisionError:
    # Code to execute if a ZeroDivisionError occurs
    print("You can't divide by zero!")
```

You can also catch multiple exceptions, use a generic `except` clause (though this is generally not recommended), and use the `finally` clause for code that should run regardless of whether an exception occurred.

````python
# Bug 5: Fixed!
def calculate_discount(price, discount_percentage):
    # Check input types
    if not (isinstance(price, (int, float)) and isinstance(discount_percentage, (int, float))):
        return "Error: Price and discount must be numbers"
    
    discount_amount = price * (discount_percentage / 100)
    final_price = price - discount_amount
    return final_price

# Test with invalid input
print(calculate_discount('fifty', 10))

# Test with valid input
print(calculate_discount(50, 10))

# Alternate solution using try/except
def calculate_discount_alt(price, discount_percentage):
    try:
        price = float(price)
        discount_percentage = float(discount_percentage)
        discount_amount = price * (discount_percentage / 100)
        final_price = price - discount_amount
        return final_price
    except (TypeError, ValueError):
        return "Error: Price and discount must be convertible to numbers"

print("\nAlternate solution:")
print(calculate_discount_alt('fifty', 10))
print(calculate_discount_alt('50', 10))  # String that can be converted to a number
````

## Exercise 3: Implementing Error Handling

Now let's practice implementing `try`/`except` blocks to handle errors gracefully. For each function below, add appropriate error handling.

````python
# 3.1: Solution
def safe_division(a, b):
    """Divide a by b and handle potential errors."""
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
        return None

# Test the function
print(safe_division(10, 2))  # Should return 5.0
print(safe_division(10, 0))  # Should handle error and return None
````

````python
# 3.2: Solution
def get_element(lst, index):
    """Get an element from a list at the specified index with error handling."""
    try:
        element = lst[index]
        return element
    except IndexError:
        print(f"Error: Index {index} is out of range for list of length {len(lst)}.")
        return None

# Test the function
my_list = ['a', 'b', 'c']
print(get_element(my_list, 1))  # Should return 'b'
print(get_element(my_list, 5))  # Should handle error and return None
````

````python
# 3.3: Solution
def read_file_content(filename):
    """Read and return the content of a file with error handling."""
    try:
        with open(filename, 'r') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' was not found.")
        return None
    except PermissionError:
        print(f"Error: You don't have permission to access '{filename}'.")
        return None
    except Exception as e:
        print(f"Error: An unexpected error occurred: {e}")
        return None

# Test the function
print(read_file_content('sample.txt'))  # Assuming this file doesn't exist

# Create a test file and test with an existing file
with open('test_file.txt', 'w') as file:
    file.write("This is a test file.")
    
print(read_file_content('test_file.txt'))
````

````python
# 3.4: Solution
def parse_positive_integer(value):
    """Parse a string to a positive integer with error handling."""
    try:
        number = int(value)
        if number < 0:
            print("Error: Number must be positive.")
            return None
        return number
    except ValueError:
        print(f"Error: '{value}' cannot be converted to an integer.")
        return None

# Test the function
print(parse_positive_integer("42"))   # Should return 42
print(parse_positive_integer("abc"))  # Should handle error and return None
print(parse_positive_integer("-5"))   # Should handle error and return None
````

## Exercise 4: Advanced Error Handling

Let's practice more complex error handling scenarios. In this exercise, you'll create functions that handle multiple types of exceptions and use finally blocks.

````python
# 4.1: Solution
def calculator(operation, a, b):
    """Calculate a result based on operation (add, subtract, multiply, divide).
    Handle all potential errors with appropriate messages."""
    try:
        # Check if a and b are numbers
        a_float = float(a)
        b_float = float(b)
        
        # Process based on operation
        if operation == 'add':
            return a_float + b_float
        elif operation == 'subtract':
            return a_float - b_float
        elif operation == 'multiply':
            return a_float * b_float
        elif operation == 'divide':
            return a_float / b_float
        else:
            raise ValueError(f"Unknown operation: {operation}")
    except TypeError:
        return f"Error: Both inputs must be numbers"
    except ValueError as e:
        if str(e).startswith("Unknown operation"):
            return f"Error: {e}"
        else:
            return f"Error: Could not convert inputs to numbers"
    except ZeroDivisionError:
        return f"Error: Cannot divide by zero"

# Test the function
print(calculator('add', 5, 3))       # Should return 8
print(calculator('divide', 10, 2))   # Should return 5.0
print(calculator('divide', 10, 0))   # Should handle division by zero
print(calculator('multiply', 4, 'x')) # Should handle type error
print(calculator('power', 2, 3))     # Should handle invalid operation
````

````python
# 4.2: Solution
def process_values(values):
    """Process a list of values by doubling numbers and capitalizing strings.
    Skip non-string, non-number values, and return the processed list."""
    processed = []
    
    for value in values:
        try:
            # Check if it's a number
            if isinstance(value, (int, float)):
                processed.append(value * 2)
            # Check if it's a string
            elif isinstance(value, str):
                processed.append(value.capitalize())
            else:
                print(f"Skipping value of type {type(value).__name__}: {value}")
        except Exception as e:
            print(f"Error processing value {value}: {e}")
    
    return processed

# Test the function
mixed_values = [10, 'hello', 3.14, [1, 2], 'python', None, 42]
print(process_values(mixed_values))
````

## Challenge: Build a Robust File Processing System

For the final challenge, you'll build a complete file processing system that demonstrates comprehensive error handling, informative error messages, and graceful failure recovery.

````python
def process_data_file(filename):
    """Process a data file with comprehensive error handling.
    
    The file should contain numbers, one per line.
    The function should:
    1. Safely open and read the file
    2. Parse each line to a number
    3. Calculate statistics (sum, average, min, max)
    4. Write the results to a new file called 'results.txt'
    5. Handle all potential errors appropriately
    
    Returns a dictionary with either the results or error information.
    """
    result = {'status': 'success', 'message': '', 'data': {}}
    numbers = []
    invalid_lines = []
    line_num = 0
    
    # Step 1: Safely open and read the file
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
    except FileNotFoundError:
        result['status'] = 'error'
        result['message'] = f"Error: File '{filename}' not found."
        return result
    except PermissionError:
        result['status'] = 'error'
        result['message'] = f"Error: No permission to read '{filename}'."
        return result
    except Exception as e:
        result['status'] = 'error'
        result['message'] = f"Error opening file: {str(e)}"
        return result
    
    # Step 2: Parse each line to a number
    for line_num, line in enumerate(lines, 1):
        line = line.strip()
        if not line:  # Skip empty lines
            continue
        
        try:
            number = float(line)
            numbers.append(number)
        except ValueError:
            invalid_lines.append((line_num, line))
    
    # Check if we have any valid numbers
    if not numbers:
        result['status'] = 'error'
        result['message'] = "No valid numbers found in the file."
        if invalid_lines:
            result['message'] += f" Invalid lines: {invalid_lines}"
        return result
    
    # Step 3: Calculate statistics
    try:
        stats = {
            'count': len(numbers),
            'sum': sum(numbers),
            'average': sum(numbers) / len(numbers),
            'min': min(numbers),
            'max': max(numbers)
        }
        result['data'] = stats
        
        # Note any invalid lines
        if invalid_lines:
            result['status'] = 'partial_success'
            result['message'] = f"Processed {len(numbers)} numbers, but found {len(invalid_lines)} invalid lines."
            result['invalid_lines'] = invalid_lines
    
    except Exception as e:
        result['status'] = 'error'
        result['message'] = f"Error calculating statistics: {str(e)}"
        return result
    
    # Step 4: Write results to a new file
    try:
        with open('results.txt', 'w') as outfile:
            outfile.write(f"Results for {filename}:\n")
            outfile.write(f"Number of valid values: {stats['count']}\n")
            outfile.write(f"Sum: {stats['sum']}\n")
            outfile.write(f"Average: {stats['average']}\n")
            outfile.write(f"Minimum: {stats['min']}\n")
            outfile.write(f"Maximum: {stats['max']}\n")
            
            if invalid_lines:
                outfile.write("\nInvalid lines found:\n")
                for line_num, line in invalid_lines:
                    outfile.write(f"Line {line_num}: {line}\n")
        
        result['message'] = f"Successfully processed {len(numbers)} numbers. Results written to 'results.txt'."
    except Exception as e:
        result['status'] = 'partial_success'
        result['message'] = f"Statistics calculated but could not write to output file: {str(e)}"
    
    return result

# Test with a good file (you can create this)
print("Testing with a good data file:")
# First, create a test file
with open('good_data.txt', 'w') as f:
    f.write("10\n20\n30\n40\n50")
print(process_data_file('good_data.txt'))

# Test with a file with some invalid data
print("\nTesting with bad data:")
with open('bad_data.txt', 'w') as f:
    f.write("10\nabc\n30\n\n40\nxyz")
print(process_data_file('bad_data.txt'))

# Test with a non-existent file
print("\nTesting with non-existent file:")
print(process_data_file('does_not_exist.txt'))
````
